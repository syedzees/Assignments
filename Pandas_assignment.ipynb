```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 'Budapest_PaRis', 'Brussels_londOn'],'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

1.Some values in the the FlightNumber column are missing. These numbers are meant to increase by 10 with each row so 10055 and 10075 need to be put in place.  Fill  in  these  missing  numbers  and  make  the  column  an  integer  column (instead of a float column).

```python
index_new=np.arange(1,df.From_To.count()+1)
df.set_index(index_new, inplace=True)
for i in np.arange(1,df.From_To.count()+1):
    if pd.isnull(df.FlightNumber.loc[i,]):
        df.loc[i,'FlightNumber'] = df.FlightNumber.loc[i-1,] + 10
        
df.FlightNumber = df.FlightNumber.astype(int)

df
```

2. The From_To column would be better as two separate columns! Split each
string on the underscore delimiter _ to give a new temporary DataFrame with
the correct values. Assign the correct column names to this temporary
DataFrame

```python
temporary_DataFrame = df.copy()

temporary_DataFrame[["From","To"]] = df.From_To.str.split("_",expand = True)

temporary_DataFrame
```

3. Notice how the capitalisation of the city names is all mixed up in this
temporary DataFrame. Standardise the strings so that only the first letter is
uppercase (e.g. "londON" should become "London".)

```python
temporary_DataFrame.From = temporary_DataFrame.From.str.capitalize()
temporary_DataFrame.To = temporary_DataFrame.To.str.capitalize()
temporary_DataFrame.From_To = temporary_DataFrame.From + "_" + temporary_DataFrame.To
temporary_DataFrame
```

4. Delete the From_To column from df and attach the temporary DataFrame
from the previous questions.

```python
df.drop("From_To", axis = 1,inplace = True)
df["From_To"] = temporary_DataFrame.From_To
df
```
